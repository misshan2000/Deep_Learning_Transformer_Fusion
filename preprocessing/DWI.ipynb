# 整理结果

```python
import os
import shutil

# 当前目录
base_dir = r'D:\pixelmedAI\function_pm\note7_diffusion\output'

# 文件分类规则
folders = {
    '_ANB.nii.gz': 'ANB',
    '_vm.nii.gz': 'vm',
    '_vs.nii.gz': 'vs'
}

# 遍历当前目录下的所有文件
for filename in os.listdir(base_dir):
    # 只处理 .nii.gz 文件（防止误操作别的文件）
    if not filename.endswith('.nii.gz') and not filename.endswith('.nii.gz.gz'):
        continue

    moved = False
    for key, folder in folders.items():
        if key in filename:
            # 创建目标文件夹
            target_dir = os.path.join(base_dir, folder)
            if not os.path.exists(target_dir):
                os.makedirs(target_dir)
            # 移动文件
            newresult = filename.split('_')[0]+'.nii.gz'
            shutil.copy(os.path.join(base_dir, filename), os.path.join(target_dir, newresult))
            moved = True
            break  # 防止同一个文件被移动多次

    if not moved:
        # 没有任何特殊后缀的文件，归类到 'others'
        target_dir = os.path.join(base_dir, 'others')
        if not os.path.exists(target_dir):
            os.makedirs(target_dir)
        shutil.copy(os.path.join(base_dir, filename), os.path.join(target_dir, filename))

print("文件已按后缀整理完毕。")
```

该实验从目标肿瘤类型的手术或穿刺活检样本中，获取高分辨组织切片并利用数字病理图像分析技术（如血管染色自动分割算法），重建代表性微血管网络。血管节点、分支和直径等结构性质参数均通过定量病理分析获得。其次，利用SpinFlowSim平台将上述微血管拓扑结构导入模型，并结合实验数据或文献资料配置血流动力学边界条件（如入口血流速度、压力及流体黏度等初值参数），开展血流仿真。SpinFlowSim基于一维稳态Navier-Stokes方程，对微血管中的血流速度及压力进行数值求解，获得全网络各点的血流分布。进一步，采用该平台集成的Bloch-Torrey方程，将微血管血流参数（血流速度分布、分支数目等）导入dMRI信号模拟，生成与实际磁共振扫描协议参数（b值序列、方向数等）相匹配的合成信号集。随后，基于合成信号构建多参数反演字典，包括血流速度均值、速度标准差以及体积加权分支指标等关键微血管参数。在本实验应用时，收集目标肿瘤患者的dMRI原始数据并进行标准预处理（含运动校正、噪声滤除等），选取与模拟信号协议一致的扫描数据。通过最小二乘法或贝叶斯反演，将体内真实信号与仿真字典拟合，逐像素估算上述微血管参数，获取空间分布图。  
均值速度（vm）：统计整个血管网络内所有血管段的流速加权平均值，反映微观血流的总体动力学水平。  
速度标准差（vs）：计算所有血管流速的标准差，描述肿瘤内血流速度的空间异质性。  
分支指数（ANB）：根据网络结构统计微血管分叉（支）的复杂程度，通过体素加权或总分支数归一化实现，定量表征肿瘤微血管结构复杂性。
